```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("amz_uk_price_prediction_dataset.csv")  # cambia el nombre si es distinto

df.head()
```

```python
category_counts = df["category"].value_counts()
category_counts.head(5)
```

```python
top_categories = category_counts.head(10)

top_categories.plot(kind="bar")
plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
top_categories.plot(kind="pie", autopct="%1.1f%%")
plt.title("Top Categories Distribution")
plt.ylabel("")
plt.show()
```

```python
mean_price = df["price"].mean()
median_price = df["price"].median()
mode_price = df["price"].mode()[0]

mean_price, median_price, mode_price
```

```python
variance = df["price"].var()
std_dev = df["price"].std()
price_range = df["price"].max() - df["price"].min()
iqr = df["price"].quantile(0.75) - df["price"].quantile(0.25)

variance, std_dev, price_range, iqr
```

```python
plt.hist(df["price"], bins=50)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
df[df["price"] < 200].hist(column="price", bins=50)
```

```python
plt.boxplot(df["price"])
plt.title("Price Boxplot")
plt.show()
```
